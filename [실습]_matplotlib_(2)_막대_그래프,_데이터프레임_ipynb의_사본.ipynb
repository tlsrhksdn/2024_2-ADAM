```python
# import
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 실습용 데이터 (이 변수들을 사용하여 그래프를 그립니다)
data = {
'Category': ['A', 'B', 'C', 'D', 'E'],
'N': [1, 2, 3, 4, 5],
'Value1': [10, 15, 8, 12, 20],
'Value2': [5, 10, 15, 7, 18],
'Value3': [18, 20, 12, 8, 15],
'Value4': [30, 22, 25, 28, 35],
'Value5': [14, 16, 9, 13, 19]
}

df = pd.DataFrame(data)
df
```

---
## 실습

#### x축에 'Category', y축에 'N'을 사용하여 간단한 막대 그래프를 그리십시오. x축은 'Category'로, y축은 'N'으로 레이블을 붙이십시오.

```python
# 예시 답
plt.bar(df['Category'], df['N'])
plt.xlabel('Category')
plt.ylabel('N')
plt.show()
```

#### y축에 'Category', x축에 'N'을 사용하여 **수평** 막대 그래프를 그리십시오. y축은 'Category'로, x축은 'N'으로 레이블을 붙이십시오.

```python
plt.bar(df['N'], df['Category'])
plt.xlabel('N')
plt.ylabel('Category')
plt.show()
```

#### x축에 'Category', y축에 'N'을 사용하여 막대 그래프를 그리십시오. x축은 'Category'로, y축은 'N'으로 레이블을 붙이고, x축 레이블을 45도 회전시키십시오.

```python
plt.bar(df['Category'], df['N'])
plt.xlabel('Category')
plt.ylabel('N')
plt.xticks(rotation=45)
plt.show()
```

#### x축에 'Category', y축에 'Value1'을 사용하여 막대 그래프를 그리고, set_hatch('//')를 사용하여 막대를 채우십시오.

```python
bars = plt.bar(df['Category'], df['Value1'])
plt.xlabel('Category')
plt.ylabel('Value1')

for bar in bars:
  bar.set_hatch('//')

plt.show()
```

#### x축에 'Category', y축에 'Value1'을 사용하여 막대 그래프를 그리고, plt.text()를 사용하여 각 막대 위에 y값을 표시하십시오.

```python
plt.xlabel('Category')
plt.ylabel('Value1')

plt.bar(df['Category'], df['Value1'])
for i, v in enumerate(df['Value1']):
  plt.text(i, v, str(v), ha='center')
```

#### x축에 'Category', y축에 'Value2'를 사용하여 막대 그래프에 그리드를 추가하십시오. 그리드 색상을 '회색', 알파를 0.2, 선 스타일을 '--'로 설정하십시오.

```python
plt.xlabel('Category')
plt.ylabel('Value2')
plt.grid(color='gray', alpha=0.2, linestyle='--')
plt.show()
```

#### x축에 'Category', y축에 'Value1'과 'Value2'를 사용하여 누적 막대 그래프를 그리십시오. 'Value1'은 아래에 있어야 합니다.

```python
plt.bar(df['Category'], df['Value1'])
plt.bar(df['Category'], df['Value2'], bottom=df['Value1'])
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

#### x축에 'Category', y축에 'Value1'과 'Value2'를 사용하여 다중 막대 그래프를 그리십시오.

```python
plt.xlabel('Category')
plt.ylabel('Value')
plt.bar(df['Category'], df['Value1'])
plt.bar(df['Category'], df['Value2'])
plt.show()
```

#### y축에 'Category', x축에 'Value3'과 'Value4'를 사용하여 다중 수평 막대 그래프를 그리십시오.

```python
plt.xlabel('Value')
plt.ylabel('Category')
plt.barh(df['Category'], df['Value3'])
plt.barh(df['Category'], df['Value4'])
plt.show()
```

#### x축에 'Category', y축에 'Value5'를 사용하여 막대 그래프를 그리십시오. y축의 범위를 0에서 30으로 설정하십시오.

```python
plt.xlabel('Category')
plt.ylabel('Value5')
plt.ylim(0, 30)
plt.bar(df['Category'], df['Value5'])
plt.show()
```

#### x축에 'Category', y축에 'Value5'를 사용하여 막대 그래프를 그리고, set_hatch('x')를 사용하여 막대를 채우십시오. 막대의 색상을 'green'으로 설정하십시오.

```python
plt.xlabel('Category')
plt.ylabel('Value5')

bars = plt.bar(df['Category'], df['Value5'], color="green")
for bar in bars:
  bar.set_hatch('x')
```

#### x축에 'Category', y축에 'Value3'을 사용하여 막대 그래프를 그리고, plt.text()를 사용하여 각 막대 위에 y값을 표시하십시오. 텍스트를 45도 회전시키십시오.

```python
plt.xlabel('Category')
plt.ylabel('Value3')

plt.bar(df['Category'], df['Value3'])
for i, v in enumerate(df['Value3']):
  plt.text(i, v, str(v), ha='center', rotation=45)
```

#### x축에 'Category', y축에 'Value1', 'Value2', 'Value3'를 사용하여 누적 막대 그래프를 그리십시오. 'Value1'은 아래에, 'Value2'는 중간에, 'Value3'는 위에 있어야 합니다.

```python
plt.xlabel('Category')
plt.ylabel('Value')

plt.bar(df['Category'], df['Value1'])
plt.bar(df['Category'], df['Value2'], bottom=df['Value1'])
plt.bar(df['Category'], df['Value3'], bottom=df['Value1'] + df['Value2'])
```

#### x축에 'Category', y축에 'Value1', 'Value2', 'Value3', 'Value4'를 사용하여 그룹화된 막대 그래프를 그리십시오.

```python
# 막대 폭 설정
bar_width = 0.2

# Category를 인덱스로 설정
df.set_index('Category', inplace=True)

# 각 막대 그리기
ax = df.plot(kind='bar', width=bar_width, position=0.5)

# x축 레이블 설정 (선택 사항)
plt.xticks(rotation=0)

# 축 레이블 및 범례 추가
plt.xlabel('Category')
plt.ylabel('Value')
plt.legend()

# 그래프 표시
plt.show()
```

#### x축에 'Category', y축에 'Value5'를 사용하여 막대 그래프를 그리고, set_hatch('/')를 사용하여 막대를 채우십시오. 막대의 색상을 '보라색'으로 설정하고, x축 레이블을 30도 회전시키십시오.

```python
plt.xlabel('Category')
plt.ylabel('Value5')

bars = plt.bar(df['Category'], df['Value5'], color="purple")

for bar in bars:
  bar.set_hatch('/')
plt.xticks(rotation=30)


plt.show()
```

#### y축에 'Category', x축에 'Value2', 'Value3', 'Value4'를 사용하여 다 수평 막대 그래프를 그리고 그래프에 그리드를 추가하십시오. 그리드 색상을 '빨강', 알파를 0.5, 선 스타일을 ':'로 설정하십시오.

```python
plt.xlabel('Value')
plt.ylabel('Category')

plt.barh(df['Category'], df['Value2'])
plt.barh(df['Category'], df['Value3'])
plt.barh(df['Category'], df['Value4'])

plt.grid(color="red", linestyle=':', alpha=0.5)
```

#### x축에 'Category', y축에 'Value2', 'Value3', 'Value4', 'Value5'를 사용하여 다중 막대 그래프를 그리십시오. 각 그래프에 다른 색상을 설정하십시오.

```python
plt.xlabel('Category')
plt.ylabel('Value')

plt.bar(df['Category'], df['Value2'], color="blue")
plt.bar(df['Category'], df['Value3'], color="green")
plt.bar(df['Category'], df['Value4'], color="red")
plt.bar(df['Category'], df['Value5'], color="purple")
```
