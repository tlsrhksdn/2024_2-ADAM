```python
# import
import matplotlib.pyplot as plt
```

```python
# 실습용 데이터
x = [1, 2, 3, 4, 5]
y1 = [10, 5, 15, 7, 20]
y2 = [15, 25, 10, 30, 20]
y3 = [12, 5, 10, 15, 8]
y4 = [18, 15, 20, 12, 25]
```

---
## 실습

#### 'x'값이 x축에 있고 'y1'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하고 그래프를 표시하세요.

```python
plt.plot(x, y1)

plt.show()
```

#### 'x'값이 x축에 있고 'y2'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하고, 그래프에 "Plot of x and y2"이라는 제목을 추가하세요.

```python
plt.plot(x, y2)

plt.title('x와 y2의 그래프')
plt.show()
```

#### 'x'값이 x축에 있고 'y3'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하고, x축을 'Time'으로, y축을 'Temperature'로 레이블링하세요.

```python
# 예시 답 (실행하여 결과를 확인해보세요)
plt.plot(x, y3)
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.show()
```

#### 'x'값이 x축에 있고 'y4'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하고, 선 그래프의 스타일을 선 너비가 2가 되도록 변경하세요.

```python
plt.plot(x, y4, linewidth=2)
plt.show()
```

#### 'x'값이 x축에 있고 'y1'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하고, 선 그래프의 스타일을 마커가 'o'가 되도록 변경하세요.

```python
plt.plot(x, y1, marker='o')
plt.show()
```

#### 'x'값이 x축에 있고 'y2'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하고, 선 그래프의 스타일을 마커가 '*', 마커 크기가 100이 되도록 수정하세요.

```python
plt.plot(x, y2, marker="*", markersize=100)
plt.show()
```

#### 'x'값이 x축에 있고 'y3'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하고, 선 그래프의 스타일을 마커가 'o', 마커 색상이 'red'가 되도록 변경하세요.

```python
plt.plot(x, y3, marker='o', color='red')
plt.show()
```

#### 'x'값이 x축에 있고 'y4'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하고, 선 그래프의 스타일을 선 스타일이 '--'가 되도록 변경하세요.

```python
plt.plot(x, y4, linestyle='--')
plt.show()
```

#### 'x'값이 x축에 있고 'y3'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하세요. 그 다음, xticks를 [1, 3, 5]로, yticks를 [10, 15, 20]으로 설정하세요.

```python
plt.plot(x, y3)
plt.xticks([1, 3, 5])
plt.yticks([10, 15, 20])
plt.show()
```

#### 'x'값이 x축에 있고 'y4'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하세요. 그 다음, "My Line"이라는 레이블로 선 그래프에 범례(Legend)를 추가하세요.

```python
plt.plot(x, y4, label='My Line')
plt.legend()
plt.show()
```

####  'x'값이 x축에 있고 'y1'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하세요. 그 다음, 그림을 'my_figure.png'로 저장하세요.

#### 'x'값이 x축에 있고 'y2'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하세요. 그 다음, 그래프의 (3, 20) 위치에 'Hello'라는 텍스트를 추가하세요.

```python
plt.plot(x, y2)
plt.text(3, 20, 'Hello')
plt.show()
```

#### (10, 5) 크기의 그림을 생성하고, 'x'값이 x축에 있고 'y3'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하세요.

- plt.figure를 사용하세요

```python
plt.plot(x, y3)
plt.figure(figsize=(10, 5))
plt.show()
```

#### 'x'값이 x축에 있고, 'y1', 'y2'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하고 그래프를 표시하세요.

```python
# 예시 답
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

#### 'x'값이 x축에 있고, 'y2', 'y4'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하고, x축을 'Time'으로, y축을 'Temperature'로 레이블링하세요.

```python
plt.plot(x, y2)
plt.plot(x, y4)
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.show()
```

####  'x'값이 x축에 있고, 'y1', 'y2', 'y3', 'y4'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하세요. 그 다음, "Line 1", "Line 2", "Line 3", "Line 4"라는 레이블로 선 그래프에 범례를 추가하세요.

```python
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.plot(x, y3, label='Line 3')
plt.plot(x, y4, label='Line 4')
```

#### 'x'값이 x축에 있고, 'y1', 'y2', 'y3', 'y4'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하세요. 선 그래프의 선 색상을 각각 다르게 설정하세요.

```python
plt.plot(x, y1, 'b')
plt.plot(x, y2, 'g')
plt.plot(x, y3, 'r')
plt.plot(x, y4, 'y')
plt.show()
```

#### 'x'값이 x축에 있고, 'y1', 'y2', 'y3', 'y4'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하세요. 선 그래프의 마커를 각각 다르게 설정하세요.

```python
plt.plot(x, y1, 'o')
plt.plot(x, y2, 's')
plt.plot(x, y3, 'x')
plt.plot(x, y4, '+')
```

#### 'x'값이 x축에 있고, 'y1', 'y2', 'y3', 'y4'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하세요. alpha를 0.3으로 설정하고 선 그래프의 선 스타일을 각각 다르게 설정하세요.

```python
plt.plot(x, y1, '-', alpha=0.3)
plt.plot(x, y2, '_', alpha=0.3)
plt.plot(x, y3, '.', alpha=0.3)
plt.plot(x, y4, '-.', alpha=0.3)
```

#### 'x'값이 x축에 있고, 'y1', 'y2', 'y3', 'y4'값이 y축에 있는 선 그래프를 그리는 Python 코드를 작성하세요. 선 그래프의 선 색상과 두께를 각각 다르게 설정하세요.

```python
plt.plot(x, y1, 'b-')
plt.plot(x, y2, 'r_')
plt.plot(x, y3, 'g.')
plt.plot(x, y4, 'y-.')
```
