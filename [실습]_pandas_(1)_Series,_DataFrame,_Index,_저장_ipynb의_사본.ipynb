#### 1부터 n까지의 n개의 자연수로 구성된 pandas 시리즈를 생성하는 파이썬 코드를 작성하십시오.

#### labels을 인덱스로 하고 values를 값으로 하는 pandas 시리즈를 생성하는 파이썬 코드를 작성하십시오.

```python
string = "this is text"
n = 5
```

```python
import pandas as pd

string = "this is text"
n = 5
pd.Series([string for i in range(n)])
```

## Ⅲ. Index

#### data 데이터를 통해 데이터프레임을 생성십시오. 인덱스는 A부터 시작하는 알파벳으로 설정하십시오.



```python
data = {
    '아이템': ['사과', '바나나', '딸기', '오렌지', '수박'],
    '가격': [1000, 500, 3000, 1500, 10000],
}
```

```python
import pandas as pd

data = {
    '아이템': ['사과', '바나나', '딸기', '오렌지', '수박'],
    '가격': [1000, 500, 3000, 1500, 10000],
}
pd.DataFrame(data, index=['A', 'B', 'C', 'D', 'E'])
```

#### 위에서 생성한 DataFrame에서 인덱스 이름을 "Item Code"로 변경해보세요.


```python
import pandas as pd

data = {
    '아이템': ['사과', '바나나', '딸기', '오렌지', '수박'],
    '가격': [1000, 500, 3000, 1500, 10000],
}
pd.DataFrame(data, index=['사과', '바나나', '딸기', '오렌지', '수박']).rename_axis('Item Code')
```

####  DataFrame을 인덱스 기준 내림차순으로 정렬해보세요.

```python
import pandas as pd

data = {
    '아이템': ['사과', '바나나', '딸기', '오렌지', '수박'],
    '가격': [1000, 500, 3000, 1500, 10000],
}

pd.DataFrame(data, index=['사과','바나나','딸기','오렌지','수박']).sort_index(ascending=False)
```

#### DataFrame의 인덱스를 초기화하되 원래 있던 인덱스는 삭제하지 않고 새로운 컬럼으로 남겨두세요.


```python
import pandas as pd

data = {
    '아이템': ['사과', '바나나', '딸기', '오렌지', '수박'],
    '가격': [1000, 500, 3000, 1500, 10000],
}

pd.DataFrame(data, index=['사과','바나나','딸기','오렌지','수박']).reset_index(drop=False)
```

## Ⅳ 저장

#### 아래의 코드를 실행하여 df 데이터프레임을 생성하고, 인덱스 이름을 sid로 설정하십시오.

```python
data = {
    'Name': ['John Doe', 'Jane Smith', 'Tom Brown', 'Alice Johnson', 'Charlie Davis'],
    'Age': [20, 25, 30, 35, 40],
}
df = pd.DataFrame(data, index=['stu1', 'stu2', 'stu3', 'stu4','stu5'])
df
```

```python
import pandas as pd

data = {
    'Name': ['John Doe', 'Jane Smith', 'Tom Brown', 'Alice Johnson', 'Charlie Davis'],
    'Age': [20, 25, 30, 35, 40],
}

df = pd.DataFrame(data, index=['stu1', 'stu2', 'stu3', 'stu4','stu5'])
df.index.name = 'sid'
df

```

#### 데이터프레임을 인덱스를 포함하여 csv파일로 저장하십시오.

```python
import pandas as pd

data = {
    'Name': ['John Doe', 'Jane Smith', 'Tom Brown', 'Alice Johnson', 'Charlie Davis'],
    'Age': [20, 25, 30, 35, 40],
}

df = pd.DataFrame(data, index=['stu1', 'stu2', 'stu3', 'stu4','stu5'])
df.index.name = 'sid'
df.to_csv('df.csv')
```

#### 데이터프레임을 인덱스를 제외하여 엑셀파일로 저장하십시오.

```python
import pandas as pd

data = {
    'Name': ['John Doe', 'Jane Smith', 'Tom Brown', 'Alice Johnson', 'Charlie Davis'],
    'Age': [20, 25, 30, 35, 40],
}

df = pd.DataFrame(data, index=['stu1', 'stu2', 'stu3', 'stu4','stu5'])
df.index.name = 'sid'

df.to_excel('df.xlsx', index=False)
```

#### csv로 저장한 데이터프레임을 앞에서부터 3개 row만 불러와 출력하십시오.

```python
import pandas as pd

data = {
    'Name': ['John Doe', 'Jane Smith', 'Tom Brown', 'Alice Johnson', 'Charlie Davis'],
    'Age': [20, 25, 30, 35, 40],
}

df = pd.DataFrame(data, index=['stu1', 'stu2', 'stu3', 'stu4','stu5'])
df.index.name = 'sid'
df.to_csv('df.csv')

pd.read_csv('df.csv', nrows=3)
```
