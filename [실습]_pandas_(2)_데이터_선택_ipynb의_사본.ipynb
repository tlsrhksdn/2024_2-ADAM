## Ⅱ. 데이터 선택(2)

- loc 또는 iloc를 사용하여 아래 문제에 답해 보세요

```python
import pandas as pd

# 사용할 데이터프레임 생성 (해당 코드박스를 실행 후 아래 문제를 풀어보세요)
df = pd.DataFrame({
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'James'],
    'Age': [28, 24, 35, 32, 30],
    'City': ['New York', 'Paris', 'Berlin', 'London', 'Sydney']
}, index=['Person 1', 'Person 2','Person 3','Person 4','Person 5'])
```

#### .loc를 사용하여 Person 1의 데이터를 선택하고 출력하세요.



```python
# 예시 정답
df.loc['Person 1']
```

#### .iloc를 사용하여 데이터프레임의 세 번째 사람의 데이터를 선택하고 출력하세요.



```python
df.iloc[[2]]
```

#### .loc를 사용하여 person4의 나이만 선택하고 출력하세요.



```python
df.loc['Person 4','Age']
```

#### .iloc를 사용하여 데이터 프레임의 두 번째 사람이 살고 있는 도시만을 선택하고 출력하세요.



```python
df.iloc[1, 2]
```

#### .loc를 사용하여 Person 3과 Person 5의 데이터를 선택하고 출력하세요.



```python
df.loc[['Person 3', 'Person 5']]
```

#### .iloc를 사용하여 데이터프레임의 첫 번째 사람과 마지막 사람의 데이터를 선택하고 출력하세요.



```python
df.iloc[[0, -1]]
```

#### .loc를 사용하여 Person 1과 Person 3의 '이름'과 '도시'를 선택하고 출력하세요.



```python
df.loc[['Person 1', 'Person 3'], ['Name', 'City']]
```

#### .iloc를 사용하여 데이터 프레임에서 홀수 위치에 있는 사람들의 모든 데이터를 선택하고 출력하세요.



```python
df.iloc[1::2]
```

#### .loc를 사용하여 Person 2, Person 4 및 Person 5의 '이름'과 '나이'를 선택하고 출력하세요.



```python
df.loc[['Person 2', 'Person 4', 'Person 5'], ['Name', 'Age']]
```

#### .iloc를 사용하여 데이터 프레임의 첫 번째, 세 번째 및 다섯 번째 사람의 이름, 나이 및 도시를 선택하고 출력하세요.



```python
df.iloc[[0, 2, 4], [0, 1, 2]]
```

## Ⅲ. 데이터 선택(조건)

- loc와 iloc를 사용하여 조건을 포함한 데이터 선택 문제를 풀어보세요.

```python
import pandas as pd

# 사용할 데이터프레임 생성 (해당 코드박스를 실행 후 아래 문제를 풀어보세요)
df = pd.DataFrame({
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'James'],
    'Age': [28, 24, 35, 32, 30],
    'City': ['New York', 'Paris', 'Berlin', 'London', 'Sydney'],
    'Gender':['Male','Female','Male','Female','Male']
}, index=['Person 1', 'Person 2','Person 3','Person 4','Person 5'])
```

```python
df
```

#### .loc를 사용하여 New York에 사는 사람들의 데이터를 선택하고 출력하세요.




```python
# 예시 정답
df.loc[df['City'] == "New York"]
```

#### .loc를 사용하여 30세보다 나이가 많은 사람들의 데이터를 선택하고 출력하세요.




```python
df.loc[df['Age'] > 30]
```

#### .loc를 사용하여 여성들의 데이터를 선택하고 출력하세요.




```python
df.loc[df['Gender'] == 'Female']
```

#### .isin()을 사용하여 Paris 또는 Berlin에 사는 사람들의 데이터를 선택하고 출력하세요.




```python
df.loc[df['City'].isin(['Paris', 'Berlin'])]
```

#### .loc를 사용하여 25세보다 나이가 많은 사람들의 이름을 선택하고 출력하세요.




```python
df.loc[df['Age'] > 25, 'Name']
```

#### .loc를 사용하여 25세보다 나이가 많은 사람들의 이름을 선택하고 출력하세요.




```python
df.loc[df['Age'] > 25, 'Name']
```

#### .isin()을 사용하여 London 또는 Sydney에 사는 사람들의 이름을 선택하고 출력하세요.



```python
df.loc[df['City'].isin(['London', 'Sydney']), 'Name']
```

#### .loc를 사용하여 30세보다 나이가 많은 사람들의 '이름'과 '나이'를 선택하고 출력하세요.




```python
df.loc[df['Age'] > 30, ['Name', 'Age']]
```

#### .isin()을 사용하여 여성 중에서 London 또는 Sydney에 사는 모든 데이터를 선택하고 출력하세요.




```python
df.loc[(df['Gender'] == 'Female') & (df['City'].isin(['London', 'Sydney']))]
```

#### .loc를 사용하여 Person 2, Person 4 및 Person 5의 '이름'과 '나이'를 선택하고 출력하세요.




```python
df.loc[['Person 2', 'Person 4', 'Person 5'], ['Name', 'Age']]
```

#### .iloc를 사용하여 데이터 프레임의 첫 번째, 세 번째 및 다섯 번째 사람의 이름, 나이 및 도시를 선택하고 출력하세요.




```python
df.iloc[[0, 2, 4], [0, 1, 2]]
```

#### .iloc를 사용하여 마지막 사람을 제외한 데이터 프레임에서 짝수 위치에 있는 사람들의 모든 데이터를 선택하고 출력하세요.




```python
df.iloc[0:-1:2]
#
```
