## Ⅲ. apply, groupby


```python
# 사용할 데이터프레임 생성 (해당 코드박스를 실행 후 아래 문제를 풀어보세요)
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'James'],
    'City': ['New York', 'London', 'London', 'Paris', 'Berlin'],
    'Age': [24, 31, 29, 18, 45],
    'Salary': [5000, 3000, 4500, 2300, 7000]
}
dataframe = pd.DataFrame(data)
```

```python
dataframe
```

#### apply 함수를 사용하여 나이가 30세 이하인 경우 각 사람을 "Junior"로 그렇지 않으면 "Senior"로 표시하는 새 열 "Seniority"를 생성하십시오.



```python
# df 데이터프레임을 사용하여 문제를 풀어보세요
df = dataframe.copy()  # 원본 데이터프레임의 복제본을 생성합니다
```

```python
# 이 코드박스에 퀴즈에 대한 답을 작성하세요
df['Seniority'] = df['Age'].apply(lambda x: 'Junior' if x <= 30 else 'Senior')
```

```python
df  # 결과물 출력
```

#### groupby 함수를 사용하여 각 도시에 사는 사람들의 평균 급여를 찾으십시오.




```python
# df 데이터프레임을 사용하여 문제를 풀어보세요
df = dataframe.copy()  # 원본 데이터프레임의 복제본을 생성합니다
```

```python
# 이 코드박스에 퀴즈에 대한 답을 작성하세요
df.groupby('City')['Salary'].mean()
```

```python
df  # 결과물 출력
```

#### groupby 함수를 사용하여 각 도시에 사는 사람들의 최대 나이를 찾으십시오.




```python
# df 데이터프레임을 사용하여 문제를 풀어보세요
df = dataframe.copy()  # 원본 데이터프레임의 복제본을 생성합니다
```

```python
# 이 코드박스에 퀴즈에 대한 답을 작성하세요
df.groupby('City')['Salary'].max()
```

```python
df  # 결과물 출력
```

#### apply 함수를 사용하여 급여가 3000 이하인 경우 각 사람을 "Low", 급여가 3001에서 6000 사이인 경우는 "Medium", 그렇지 않으면 "High"로 표시하는 새 열 "Salary Category"를 생성하십시오.



```python
# df 데이터프레임을 사용하여 문제를 풀어보세요
df = dataframe.copy()  # 원본 데이터프레임의 복제본을 생성합니다
```

```python
# 이 코드박스에 퀴즈에 대한 답을 작성하세요
df['Salary Category'] = df['Salary'].apply(lambda x: 'Low' if x <= 3000 else ('Medium' if x <= 6000 else 'High'))
```

```python
df  # 결과물 출력
```

#### groupby 함수를 사용하여 각 도시에 사는 사람들의 총 급여를 찾으십시오.


```python
# df 데이터프레임을 사용하여 문제를 풀어보세요
df = dataframe.copy()  # 원본 데이터프레임의 복제본을 생성합니다
```

```python
# 이 코드박스에 퀴즈에 대한 답을 작성하세요
df.groupby('City')['Salary'].sum()
```
