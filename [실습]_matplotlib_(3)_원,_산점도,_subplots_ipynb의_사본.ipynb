```python
# import
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 실습용 데이터 (이 변수들을 사용하여 그래프를 그립니다)
data = {
'Category': ['A', 'B', 'C', 'D', 'E'],
'N': [1, 2, 3, 4, 5],
'Value1': [10, 15, 8, 12, 20],
'Value2': [5, 10, 15, 7, 18],
'Value3': [18, 20, 12, 8, 15],
'Value4': [30, 22, 25, 28, 35],
'Eql': [1, 1, 1, 1, 1]
}

df = pd.DataFrame(data)
df
```

---
## 실습

#### 데이터의 'N' 열을 사용하여 파이 차트를 그려보세요. 'Category' 열을 라벨로 사용하세요.

```python
plt.pie(df['N'], labels=df['Category'])
plt.show()
```

#### 'Value1' 열을 사용하여 파이 차트를 그려보세요. 'Category'를 라벨로 사용하고 각 섹션의 백분율을 소수점 1자리로 표시하세요.

```python
# 예시 답
plt.pie(df['Value1'], labels=df['Category'], autopct='%1.1f%%')
plt.show()
```

#### 'Value1'과 'Value2'를 사용하여 산점도를 그려보세요. 'Value3'를 점의 크기로 사용하세요.

```python
scatter = plt.scatter(df['Value1'], df['Value2'], s=df['Value3'])
plt.show()
```

#### 'Value1'과 'Value2'를 사용하여 산점도를 그려보세요. 'Value3'를 점의 크기로, 'Value4'를 점의 색상으로 사용하세요.

```python
scatter = plt.scatter(df['Value1'], df['Value2'], s=df['Value3'], c=df['Value4'])
plt.show()
```

#### 'Value1' 열을 사용하여 파이 차트를 그려보세요. 'Category'를 라벨로 사용하고 각 섹션의 백분율을 표시하세요. 파이는 반시계 방향으로 그려야 합니다.

```python
plt.pie(df['Value1'], labels=df['Category'], autopct="%1.1f%%", counterclock=True)
```

#### 'Value1'과 'Value2'를 사용하여 산점도를 그려보세요. 'Value3'를 점의 크기로, 'Value4'를 색상으로 사용하세요. 플롯에 colorbar를 추가하세요.

```python
scatter = plt.scatter(df['Value1'], df['Value2'], s=df['Value3'], c=df['Value4'])
plt.colorbar(scatter)
```

#### 2행의 서브플롯을 만드세요. 첫 번째 행에서는 'N'과 'Category'를 사용하여 파이 차트를, 두 번째 행에서는 'Value1'과 'Value2'를 사용하여 산점도를 그려보세요.

```python
# 예시 답
fig, ax = plt.subplots(2)
ax[0].pie(df['N'], labels=df['Category'])
ax[1].scatter(df['Value1'], df['Value2'])
plt.show()
```

#### 2행의 서브플롯을 만드세요. 첫 번째 행에서는 'N'과 'Category'를 사용하여 파이 차트를 그려보세요. 두 번째 행에서는 'Value1'과 'Value2'를 사용하여 산점도를 그려보세요. 서브플롯의 배경색을 'lightblue'로 설정하세요.

```python
flg, ax = plt.subplots(2)
ax[0].pie(df['N'], labels=df['Category'])
ax[1].scatter(df['Value1'], df['Value2'])

for i in range(2):
    ax[i].set_facecolor('lightblue')

plt.show()
```

#### 'Value1'과 'Value2'를 사용하여 산점도를 그려보세요. 'Value3'를 점의 크기로, 'Value4'를 색상으로 사용하세요. 알파값을 0.5로 설정하세요.

```python
scatter = plt.scatter(df['Value1'], df['Value2'], s=df['Value3'], c=df['Value4'], alpha=0.5)
plt.show()
```

#### 2행의 서브플롯을 만드세요. 첫 번째 행에서는 'N'과 'Category'를 사용하여 파이 차트를 그려보세요. 두 번째 행에서는 'Value1'과 'Value2'를 사용하여 산점도를 그려보세요. 산점도에 격자를 추가하세요.

```python
fig, ax = plt.subplots(2)
ax[0].pie(df['N'], labels=df['Category'])
ax[1].scatter(df['Value1'], df['Value2'])
ax[1].grid(True)

plt.show()
```

#### 'Value2' 열을 사용하여 파이 차트를 그려보세요. 'Category'를 라벨로 사용하고 각 섹션의 백분율을 표시하세요. 파이는 90도 각도에서 시작해 그려야 합니다.

```python
plt.pie(df['Value2'], labels=df['Category'], startangle=90)
```

#### 2x2의 서브플롯을 만드세요. 첫 번째 행에서는 'Value1', 'Value2'와 'Category'를 사용하여 2개의 파이 차트를 그려보세요. 두 번째 행에서는 'Value3'과 'Value4'를 사용한 산점도를 하나, 'N'과 'Value4'를 사용한 선 그래프를 하나 그려보세요. 서브플롯의 배경색을 'yellow'로 설정하세요.

```python
fig, ax = plt.subplots(2, 2)

ax[0, 0].pie(df['Value1'], labels=df['Category'])
ax[0, 1].pie(df['Value2'], labels=df['Category'])

ax[1, 0].pie(df['Value3'], labels=df['Category'])
ax[1, 1].pie(df['Value4'], labels=df['Category'])

for i in range(2):
    for j in range(2):
        ax[i, j].set_facecolor('yellow')
```
